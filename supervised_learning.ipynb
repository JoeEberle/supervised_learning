## Step 1 - Import supervised learning libraries 

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import svm 
```

## Step 2 - Load the dataset 

```python
iris = datasets.load_iris()  
```

## Step 3 - Split the data set into training and Test 

```python
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)
```

## Step 4 - Establish a Classifier 

```python
clf = svm.SVC(kernel='linear') 
```

## Step 5 - Train the Classifier using the training sets

```python
clf.fit(x_train, y_train)
```

## Step 6 - Predict the response for the test dataset 

```python
y_pred = clf.predict(x_test)
```

```python
import matplotlib.pyplot as plt
from matplotlib.patheffects import withStroke

def create_vertical_sign(word, font_size=100, font_color='lime', background_color='black', glow_color='lime', filename=None):
    # Create a figure and axis with no frame
    word = word.replace("_"," ").title()
    fig, ax = plt.subplots(figsize=(8, 2))
    ax.axis('off')

    # Set background color
    ax.set_facecolor(background_color)

    # Render the text with a glow effect
    text = ax.text(0.2, 0.8, word, ha='center', va='center', fontsize=font_size, color=font_color,
                   path_effects=[withStroke(linewidth=2, foreground=glow_color)])

    # Save the figure as PNG if filename is provided
    if filename:
        plt.savefig(filename, bbox_inches='tight', pad_inches=0, transparent=True)

    plt.show()

# Example usage:
create_vertical_sign("Automated Sign Painting", font_size=35, font_color='gray', background_color='blue', glow_color='black', filename="sign_NEON.png")
create_vertical_sign("cancer_treatment_side_effects", font_size=35, font_color='black', background_color='black', glow_color='green', filename="sign_NEON.png")
```

```python
import matplotlib.pyplot as plt
from matplotlib.patheffects import withStroke

def create_stacked_sign(phrase, font_size=100, font_color='lime', background_color='black', glow_color='lime', filename=None):
    # Split the phrase into individual words
    phrase = phrase.replace("_"," ").title()
    words = phrase.split()

    # Calculate total height of stacked words
    total_height = len(words) * font_size * 1.9  # Adjust spacing as needed

    # Create a figure and axis with appropriate size
    fig, ax = plt.subplots(figsize=(8, total_height / 100))  # Divide total height by 100 for inches

    # Set background color
    ax.set_facecolor(background_color)

    # Render the words with glow effect and center them vertically
    for i, word in enumerate(words):
        text_y = (total_height - (i * font_size * 1.8)) / total_height  # Calculate y position
        text = ax.text(0.5, text_y, word, ha='center', va='center', fontsize=font_size, color=font_color,
                       path_effects=[withStroke(linewidth=2, foreground=glow_color)])

    # Hide axes
    ax.axis('off')

    # Save the figure as PNG if filename is provided
    if filename:
        plt.savefig(filename, bbox_inches='tight', pad_inches=0, transparent=True)

    plt.show()

# Example usage:
phrase = "cancer_treatment_side_effects"
create_stacked_sign(phrase, font_size=40, font_color='lightgray', background_color='black', glow_color='black', filename="sign_NEON.png")
create_vertical_sign(phrase, font_size=35, font_color='orange', background_color='orange', glow_color='red', filename="sign_NEON.png")
create_vertical_sign(phrase, font_size=55, font_color='purple', background_color='orange', glow_color='lightgray', filename="sign_NEON.png")
```

```python
def present_dataframe_as_a_list(df, term_column_number = 0, definition_column_number = 1,  output="print"):
    for index, row in df.iterrows():
        if output=="print" or output=="printsay":
            print(f"{index + 1}. {row[df.columns[term_column_number]]} - {row[df.columns[definition_column_number]]}")
        if output=="speech" or output=="printsay":
            tc.say(f"{index + 1}. {row[df.columns[term_column_number]]} - {row[df.columns[definition_column_number]]}")            
    return  
```
